# Creating Portfolio's

```python
pol_counter=0
#no_of_policies_in_portfolio_list=[1000,3000,5000]
no_of_policies_in_portfolio_list=[5000,10000,15000]
for no_of_policies_in_portfolio in no_of_policies_in_portfolio_list:
    for every_iteration in range(kz):
        for each_percentage in policy_loss_percentages:
            namar='Portfolio_'+str(pol_counter)+".csv" 
            pol_counter=pol_counter+1
            portfolio_neg_portion = negative_.sample(int(no_of_policies_in_portfolio * (each_percentage/100)))
            portfolio_pos_portion = postivie_.sample(int(no_of_policies_in_portfolio - (int(no_of_policies_in_portfolio * (each_percentage/100)))))
            new_portfolio = pd.concat([portfolio_neg_portion ,portfolio_pos_portion ])
            new_portfolio.to_csv(os.path.join('training_portfolios',namar))
```

```python
def Extract_data_from_portfolio(portfolio_name):
    pre="./training_portfolios/"
    df_temp=pd.read_csv(pre+portfolio_name)
    processed_data=dict()
    processed_data['ID']= portfolio_name
    processed_data['Vehicle_Territory_Avg']=np.mean(df_temp['Vehicle_Territory'])
    processed_data['Vehicle_Territory_std']=np.std(df_temp['Vehicle_Territory'])
    processed_data['Vehicle_Make_Year_Avg']=np.mean(df_temp['Vehicle_Make_Year'])
    processed_data['Vehicle_Make_Year_std']=np.std(df_temp['Vehicle_Make_Year'])
    processed_data['Vehicle_Performance_Avg']=np.mean(df_temp['Vehicle_Performance'])
    processed_data['Vehicle_Performance_Std']=np.std(df_temp['Vehicle_Performance'])
    processed_data['Vehicle_New_Cost_Amount_Avg']=np.mean(df_temp['Vehicle_New_Cost_Amount'])
    processed_data['Vehicle_New_Cost_Amount_Std']=np.std(df_temp['Vehicle_New_Cost_Amount'])
    processed_data['Vehicle_Number_Of_Drivers_Assigned_Avg']=np.mean(df_temp['Vehicle_Number_Of_Drivers_Assigned'])
    processed_data['Vehicle_Number_Of_Drivers_Assigned_Std']=np.std(df_temp['Vehicle_Number_Of_Drivers_Assigned'])
    processed_data['Vehicle_Usage_Avg']=np.mean(df_temp['Vehicle_Usage'])
    processed_data['Vehicle_Usage_Std']=np.std(df_temp['Vehicle_Usage'])
    processed_data['Vehicle_Miles_To_Work_Avg']=np.mean(df_temp['Vehicle_Miles_To_Work'])
    processed_data['Vehicle_Miles_To_Work_Std']=np.std(df_temp['Vehicle_Miles_To_Work'])
    processed_data['Vehicle_Annual_Miles_Avg']=np.mean(df_temp['Vehicle_Annual_Miles'])
    processed_data['Vehicle_Annual_Miles_Std']=np.std(df_temp['Vehicle_Annual_Miles'])
    processed_data['Vehicle_Anti_Theft_Device_0']=((df_temp[df_temp['Vehicle_Anti_Theft_Device']==0].shape[0]) / df_temp.shape[0])
    processed_data['Vehicle_Anti_Theft_Device_1']=((df_temp[df_temp['Vehicle_Anti_Theft_Device']==1].shape[0]) / df_temp.shape[0])
    processed_data['Vehicle_Anti_Theft_Device_2']=((df_temp[df_temp['Vehicle_Anti_Theft_Device']==2].shape[0]) / df_temp.shape[0])
    processed_data['Vehicle_Anti_Theft_Device_3']=((df_temp[df_temp['Vehicle_Anti_Theft_Device']==3].shape[0]) / df_temp.shape[0])
    processed_data['Vehicle_Age_In_Years_Avg']=np.mean(df_temp['Vehicle_Age_In_Years'])
    processed_data['Vehicle_Age_In_Years_Std']=np.std(df_temp['Vehicle_Age_In_Years'])
    processed_data['Vehicle_Physical_Damage_Limit_Avg']=np.mean(df_temp['Vehicle_Physical_Damage_Limit'])
    processed_data['Vehicle_Physical_Damage_Limit_Std']=np.std(df_temp['Vehicle_Physical_Damage_Limit'])
    processed_data['Vehicle_Collision_Coverage_Deductible_Avg']=np.mean(df_temp['Vehicle_Collision_Coverage_Deductible'])
    processed_data['Vehicle_Collision_Coverage_Deductible_Std']=np.std(df_temp['Vehicle_Collision_Coverage_Deductible'])
    processed_data['Driver_Male_Female_Ratio']=(df_temp['Driver_Total_Male'].sum() / df_temp['Driver_Total_Female'].sum())
    processed_data['Driver_Single_Married_Ratio']=(df_temp['Driver_Total_Single'].sum() / df_temp['Driver_Total_Married'].sum())
    processed_data['Driver_Minimum_Age_Avg']=np.mean(df_temp['Driver_Minimum_Age'])
    processed_data['Driver_Minimum_Age_Std']=np.std(df_temp['Driver_Minimum_Age'])
    processed_data['Driver_Maximum_Age_Avg']=np.mean(df_temp['Driver_Maximum_Age'])
    processed_data['Driver_Maximum_Age_Std']=np.std(df_temp['Driver_Maximum_Age'])
    processed_data['Vehicle_Driver_Points_Avg']=np.mean(df_temp['Vehicle_Driver_Points'])
    processed_data['Vehicle_Driver_Points_Std']=np.std(df_temp['Vehicle_Driver_Points'])
    processed_data['Driver_Total_Senior_Teenager_Ratio'] = (df_temp['Driver_Total_Senior_Ages_65_69'].sum() / df_temp['Driver_Total_Teenager_Age_15_19'].sum())
    processed_data['Driver_Total_Upper_Senior_Middle_Adult_Ratio'] = (df_temp['Driver_Total_Upper_Senior_Ages_70_plus'].sum() / df_temp['Driver_Total_Middle_Adult_Ages_40_49'].sum())
    processed_data['Vehicle_Days_Per_Week_Driven_Avg'] = np.mean(df_temp['Vehicle_Days_Per_Week_Driven'])
    processed_data['Vehicle_Days_Per_Week_Driven_std'] = np.std(df_temp['Vehicle_Days_Per_Week_Driven'])
    processed_data['Vehicle_Physical_Damage_Med_pay_Ratio'] = (df_temp['Vehicle_Physical_Damage_Limit'].sum() / df_temp['Vehicle_Med_Pay_Limit'].sum())
    processed_data['Vehicle_Symbol_Avg'] = np.mean(df_temp['Vehicle_Symbol'])
    processed_data['Vehicle_Symbol_std'] = np.std(df_temp['Vehicle_Symbol'])
    processed_data['Annual_Premium_Avg'] = np.mean(df_temp['Annual_Premium'])
    processed_data['Annual_Premium_std']= np.std(df_temp['Annual_Premium'])
    processed_data['ln_LR'] = np.log((df_temp['Loss_Amount'].sum() / df_temp['Annual_Premium'].sum()))

    return processed_data
```
